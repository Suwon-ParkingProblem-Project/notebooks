```python
import pandas as pd

df = pd.read_csv("수원시_주정차단속현황.csv", encoding="cp949")
df.head(5)
```

# 권선구

```python
# 시군구명 컬럼이 권선구경제교통과인 것만 별도로 추출
df_gy = df[df["시군구명"] == "권선구경제교통과"]

# 단속구분 컬럼 유니크값 확인
df_gy["단속구분"].unique()
```

```python
# 단속구분 컬럼이 고정형CCTV(static)인 경우만 추출
df_gy_cctv_s = df_gy[df_gy["단속구분"] == "고정형CCTV"]
df_gy_cctv_s.head(5)
```

```python
# 단속구분 컬럼이 주행형CCTV(drive)인 경우만 추출
df_gy_cctv_d = df_gy[df_gy["단속구분"] == "주행형CCTV"]
df_gy_cctv_d.head(5)
```

```python
# 단속구분 컬럼이 국민신문고, 모바일(개별시스템), 민원(공익제보), 안전신문고인 경우를 민원(people) 컬럼으로 추출
df_gy_p = df_gy[df_gy["단속구분"].isin(["국민신문고", "모바일(개별시스템)", "민원(공익제보)", "안전신문고"])]
df_gy_p.head(5)
```

```python
# 단속구분 컬럼이 수기보행등록, PDA인 경우를 단속원(official) 컬럼으로 추출
df_gy_o = df_gy[df_gy["단속구분"].isin(["수기보행등록", "PDA"])]
df_gy_o.head(5)
```
