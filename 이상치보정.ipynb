```python
import pandas as pd
import numpy as np
import koreanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import statsmodels.api as sm
```

```python
df = pd.read_csv("Outlier.csv", index_col = False, encoding = 'cp949')
```

```python
df.set_index("행정동", inplace = True)
```

```python
fig, axes = plt.subplots(5, 1, figsize=(20,20))
fig.suptitle('단속건수 방식별 박스플롯', fontsize=40)
#plt.tight_layout()

for ax,feature in zip(axes.flatten(),df):
    sns.boxenplot(data = df, x = feature, ax=ax, color='#f15354')
plt.show()
```

```python
df_2 = np.log(df[['고정형', '주행형', '보행', '국민신고', '전체']])
```

```python
fig, axes = plt.subplots(5, 1, figsize=(20,20))
fig.suptitle('단속건수 방식별 박스플롯', fontsize=40)
#plt.tight_layout()

for ax,feature in zip(axes.flatten(),df_2):
    sns.boxenplot(data = df_2, x = feature, ax=ax, color='#f15354')
plt.show()
```
